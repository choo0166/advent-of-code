## Part 1

```python
def findSumOfPartNumbers(input: np.ndarray) -> int:
    rows, cols = input.shape
    symbols = "!#$%&'()*+,-/:;<=>?@[\]^_`{|}~"
    partNums = []

    def isPartNumber(indexes):
        """
        Checks for adjacent symbols in each index of the grid that are digits
        """
        for x, y in indexes:
            for i, j in [
                (x-1, y),
                (x+1, y),
                (x, y-1),
                (x, y+1),
                (x-1, y-1),
                (x-1, y+1),
                (x+1, y-1),
                (x+1, y+1)
                ]:
                if i < 0 or i >= rows or j < 0 or j >= cols:
                    continue
                elif input[i, j] in symbols:
                    return True
        return False

    for i in range(rows):
        num, indexes = "", []
        for j in range(cols):
            if input[i, j].isdigit():
                num += input[i, j]
                indexes.append((i, j))
            else:
                # we hit a non-digit, either we have a part number or not
                # in either case, we reset the number string and indexes
                # to check
                if isPartNumber(indexes):
                    # print(indexes)
                    partNums.append(int(num))
                num = ""
                indexes.clear()

        # checks for a part number at the last column
        if isPartNumber(indexes):
            partNums.append(int(num))

    print(partNums)
    return sum(partNums)
```

## Part 2

```python
from operator import itemgetter
from functools import reduce

def findSumOfGearRatios(input: np.ndarray) -> int:
    rows, cols = input.shape
    potential = np.argwhere(input == "*")   # indexes of potential gears
    ans = 0

    def getPartNumbers(x, y):
        """
        Helper method to return all part numbers from an index with * symbol
        """
        nums = set()
        for i, j in [
                (x-1, y),
                (x+1, y),
                (x, y-1),
                (x, y+1),
                (x-1, y-1),
                (x-1, y+1),
                (x+1, y-1),
                (x+1, y+1)
                ]:
            if i < 0 or i >= rows or j < 0 or j >= cols:
                continue
            elif input[i, j].isdigit():
                start = end = j
                # scan from the left and right of the adjacent index to
                # find all consecutive set of digits that form the part
                # number
                for c in range(j-1, -1, -1):
                    if input[i, c].isdigit():
                        start -= 1
                    else: break

                for c in range(j+1, cols):
                    if input[i, c].isdigit():
                        end += 1
                    else: break

                nums.add(int("".join(input[i, start:end+1])))
        return nums

    # print(potential)
    for x, y in potential:
        neighbour_nums = getPartNumbers(x, y)
        if len(neighbour_nums) == 2:
            ans += reduce(lambda x, y: x*y, neighbour_nums)

    return ans
```
