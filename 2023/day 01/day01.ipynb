## Part 1

```python
from typing import List

def findTotalCalibrationValueP1(input: List[str]) -> int:
    ans = 0
    for line in input:
        digits = []
        for char in line.strip():
            if char.isdigit(): digits.append(char)

        if len(digits) > 1:
            ans += int(digits[0] + digits[-1])
        elif len(digits) == 1:
            # only 1 digit so first and last digit are the same
            ans += int(digits[0]*2)

    return ans
```

## Part 2

```python
test = """
two1nine
eightwothree
abcone2threexyz
xtwone3four
4nineeightseven2
zoneight234
7pqrstsixteen
""".strip().splitlines()
```

```python
def findTotalCalibrationValueP2(input: List[str]) -> int:
    # map of starting letters of digits to possible digits
    starts = {
        "o": [("one", "1")],
        "t": [("two", "2"), ("three", "3")],
        "f": [("four", "4"), ("five", "5")],
        "s": [("six", "6"), ("seven", "7")],
        "e": [("eight", "8")],
        "n": [("nine", "9")]
    }
    ans = 0
    for line in input:
        digits = []
        for i, char in enumerate(line.strip()):
            if char in starts:
                for letter, digit in starts[char]:
                    if line[i:i+len(letter)] == letter:
                        digits.append(digit)
                        break
            elif char.isdigit(): digits.append(char)

        if len(digits) > 1:
            ans += int(digits[0] + digits[-1])
        elif len(digits) == 1:
            # only 1 digit so first and last digit are the same
            ans += int(digits[0]*2)

    return ans
```

```python
assert findTotalCalibrationValueP2(test) == 281, "Fails test case!"
findTotalCalibrationValueP2(data)
```
