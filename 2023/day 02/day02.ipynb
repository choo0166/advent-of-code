## Part 1

```python
from typing import List

def findSumOfPossibleIds(input: List[str]) -> int:
    bag = {
        "red": 12,
        "green": 13,
        "blue": 14
    }
    ans = 0
    for line in input:
        key, subsets = line.strip().split(": ")
        reveals = [x.split(", ") for x in subsets.split("; ")]
        possible = True
        for reveal in reveals:
            for outcome in reveal:
                num, color = outcome.split()
                if int(num) > bag[color]:
                    possible = False
                    break
            if not possible:
                break

        if possible: ans += int(key.split()[1])

    return ans
```

```python
test = """
Game 1: 3 blue, 4 red; 1 red, 2 green, 6 blue; 2 green
Game 2: 1 blue, 2 green; 3 green, 4 blue, 1 red; 1 green, 1 blue
Game 3: 8 green, 6 blue, 20 red; 5 blue, 4 red, 13 green; 5 green, 1 red
Game 4: 1 green, 3 red, 6 blue; 3 green, 6 red; 3 green, 15 blue, 14 red
Game 5: 6 red, 1 blue, 3 green; 2 blue, 1 red, 2 green
""".strip().splitlines()
```

```python
assert findSumOfPossibleIds(test) == 8, "Fails test case!"
```

```python
findSumOfPossibleIds(data)
```

## Part 2

```python
from collections import defaultdict
from functools import reduce

def findSumOfPowers(input: List[str]) -> int:
    ans = 0
    for line in input:
        key, subsets = line.strip().split(": ")
        reveals = [x.split(", ") for x in subsets.split("; ")]
        bag = defaultdict(int)
        for reveal in reveals:
            for outcome in reveal:
                num, color = outcome.split()
                # Update maximum number of each type of cube in each draw
                if int(num) > bag[color]:
                    bag[color] = int(num)

        # print(bag)
        ans += reduce(lambda x, y: x*y, bag.values())

    return ans
```

```python
assert findSumOfPowers(test) == 2286, "Fails test case!"
```

```python
findSumOfPowers(data)
```
