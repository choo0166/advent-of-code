```python
import numpy as np

data = open(
    '/content/drive/MyDrive/Colab Notebooks/AOC/2022/day8.txt',
    'r').readlines()

mat = []
for line in data:
    row = list(line.rstrip())
    mat.append(list(map(int, row)))

mat = np.array(mat, dtype = np.int8)
```

## Part 1

```python
def findNumVisibleTrees(input: np.ndarray) -> int:
    rows, cols = input.shape
    num_visible = (rows * 4) - 4  # count trees on edges as visible
    def is_visible(x, y):
        # Check along leftside of row of coordinate
        if input[x, :y].max() < input[x][y]:
            return True
        # Check along rightside of row of coordinate
        elif input[x, y+1:].max() < input[x][y]:
            return True
        # Check along top of column of coordinate
        elif input[:x, y].max() < input[x][y]:
            return True
        # Check along bottom of column of coordinate
        elif input[x+1:, y].max() < input[x][y]:
            return True
        else:
            return False

    for i in range(1, rows - 1):
        for j in range(1, cols - 1):
            if is_visible(i, j): num_visible += 1

    return num_visible
```

```python
findNumVisibleTrees(mat)
```

## Part 2

```python
def findMaxScenicScore(input: np.ndarray) -> int:
    rows, cols = input.shape
    max_score = 0
    def compute_scenic_score(x, y):
        if x == 0 or x == rows-1 or y == 0 or y == cols-1:
            # trees along edges have 0 viewing distance
            return 0
        left, right, top, bottom = 0, 0, 0, 0   # viewing distances
        # Check along leftside of row of coordinate
        for height in reversed(input[x, :y]):
            if height < input[x][y]: left += 1
            else: left += 1; break
        # Check along rightside of row of coordinate
        for height in input[x, y+1:]:
            if height < input[x][y]: right += 1
            else: right += 1; break
        # Check along top of column of coordinate
        for height in reversed(input[:x, y]):
            if height < input[x][y]: top += 1
            else: top += 1; break
        # Check along bottom of column of coordinate
        for height in input[x+1:, y]:
            if height < input[x][y]: bottom += 1
            else: bottom += 1; break
        return left*right*top*bottom

    for i in range(rows):
        for j in range(cols):
            scenic_score = compute_scenic_score(i, j)
            if scenic_score > max_score: max_score = scenic_score

    return max_score
```

```python
findMaxScenicScore(mat)
```
