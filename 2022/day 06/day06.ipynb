## Part 1

```python
def findNumCharsBeforeMarkerP1(input: str) -> int:
    for i in range(len(input)):
        seq = input[i:i+4]
        if len(set(seq)) == 4:
            return i + 4    # i + 1 characters processed at start of packet
    return None
```

```python
print(findNumCharsBeforeMarkerP1(data))
```

## Part 2

```python
def findNumCharsBeforeMarkerP2(input: str, markerSize: int) -> int:
    for i in range(len(input)):
        seq = input[i:i+markerSize]
        if len(set(seq)) == markerSize:
            return i + markerSize
    return None
```

```python
print(findNumCharsBeforeMarkerP2(data, 14))
```
