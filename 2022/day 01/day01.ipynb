## Part 1

```python
from typing import List

def findMaxCal(input: List[str]) -> int:
    ans = 0
    for record in input:
        total = sum([int(n) for n in record.split("\n")])
        if total > ans:
            ans = total
    return ans
```

```python
findMaxCal(data)
```

## Part 2

```python
def findTop3ElfsCal(input: List[str]) -> int:
    calories = []
    for record in input:
        total = sum([int(n) for n in record.split("\n")])
        calories.append(total)
    calories.sort(reverse=True)
    return sum(calories[:3])
```

```python
findTop3ElfsCal(data)
```
