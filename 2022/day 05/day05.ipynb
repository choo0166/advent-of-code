```python
from collections import defaultdict

# Build map of stack -> crates
stackMap = defaultdict(list)
lines = data[0].split("\n")

"""
Iterate over the list of lines in reverse to append crates in proper stacking
order, note that each crate occupies 3 characters from the starting index of
each stack and stack has starting index in multiples of 4. A non-existent crate
is marked by 3 whitespaces.
"""
for line in lines[::-1][1:]:
    for i in range(0, len(line), 4):
        stack = (i // 4) + 1
        crate = line[i: i+3]
        if crate != '   ':
            stackMap[stack].append(crate)

for k, v in stackMap.items():
    print(k ,v)
```

## Part 1

```python
from typing import List, Dict

def findCrateAssignmentsP1(moves: str, stacks: Dict[int, List[str]]) -> str:
    for line in moves.split("\n"):
        l = line.rstrip().split(" ")
        num, fr, to = int(l[1]), int(l[3]), int(l[-1])
        stacks[to].extend(reversed(stacks[fr][-num:])) # append new crates
        stacks[fr] = stacks[fr][:len(stacks[fr]) - num] # remove appended crates
    s = ""
    for v in stacks.values():
        s += (v[-1][1:-1])
    return s
```

```python
findCrateAssignmentsP1(data[1], stackMap)
```

## Part 2

```python
def findCrateAssignmentsP2(moves: str, stacks: Dict[int, List[str]]) -> str:
    for line in moves.split("\n"):
        l = line.rstrip().split(" ")
        num, fr, to = int(l[1]), int(l[3]), int(l[-1])
        stacks[to].extend(stacks[fr][-num:]) # append new crates
        stacks[fr] = stacks[fr][:len(stacks[fr]) - num] # remove appended crates
    s = ""
    for v in stacks.values():
        s += (v[-1][1:-1])
    return s
```

```python
findCrateAssignmentsP2(data[1], stackMap)
```
