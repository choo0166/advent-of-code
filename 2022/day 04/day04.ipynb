## Part 1

```python
from typing import List
from operator import itemgetter

def findNumOverlappingAssignmentsP1(input: List[str]) -> int:
    ans = 0
    for line in input:
        pairs = line.strip().split(",")
        asn = [tuple(list(map(int, p.split("-")))) for p in pairs]
        asn.sort(key = itemgetter(0))
        # Checks if 2nd assignment is fully contained in first
        if asn[-1][0] <= asn[0][1] and asn[-1][1] <= asn[0][1]:
            ans += 1
        # If starts are tied, check if 1st assignment is fully contained in second
        elif asn[0][0] == asn[-1][0] and asn[0][1] <= asn[-1][1]:
            ans += 1

    return ans
```

```python
findNumOverlappingAssignmentsP1(assignments)
```

## Part 2

```python
def findNumOverlappingAssignmentsP2(input: List[str]) -> int:
    ans = 0
    for line in input:
        pairs = line.strip().split(",")
        asn = [tuple(list(map(int, p.split("-")))) for p in pairs]
        asn.sort(key = itemgetter(0))
        if asn[-1][0] <= asn[0][1]:
            ans += 1

    return ans
```

```python
findNumOverlappingAssignmentsP2(assignments)
```
