```python
import numpy as np
data = open(
    '/content/drive/MyDrive/Colab Notebooks/AOC/2021/day9.txt',
    'r')

mat = []

for line in data.readlines():
    arr = list(line.rstrip())
    mat.append(list(map(int, arr)))

mat = np.array(mat, dtype = np.int8)
```

## Part 1

```python
def findRiskLow(mat: np.ndarray) -> int:
    rows, cols = mat.shape[0], mat.shape[1]
    risk = 0

    def isLowPoint(i, j):
        adj = []
        # look at points on top, bottom, left and right
        for x, y in [(i-1, j), (i+1, j), (i, j-1), (i, j+1)]:
            if x < 0 or x >= rows or y < 0 or y >= cols:
                continue
            adj.append(mat[x][y])

        return mat[i][j] < min(adj)

    for i in range(rows):
        for j in range(cols):
            if isLowPoint(i, j): risk += (1 + mat[i,j])

    return risk
```

## Part 2

```python
from functools import reduce

def findTop3Basins(mat: np.ndarray) -> int:
    """
    Idea: Starting from each low point, call dfs on adjacent points (as long as
    it is larger) until we hit the highest point(9), keeping track of points
    in our path that form the basin
    """
    rows, cols = mat.shape[0], mat.shape[1]
    sizes = []
    def isLowPoint(i, j):
        adj = []
        # look at points on top, bottom, left and right
        for x, y in [(i-1, j), (i+1, j), (i, j-1), (i, j+1)]:
            if x < 0 or x >= rows or y < 0 or y >= cols:
                continue
            adj.append(mat[x][y])

        return mat[i][j] < min(adj)

    def findLowPoints(mat):
        ans = []
        for i in range(rows):
            for j in range(cols):
                if isLowPoint(i, j): ans.append((i, j))
        return ans

    def dfs(i, j, visited):
        for x, y in [(i-1, j), (i+1, j), (i, j-1), (i, j+1)]:
            if x < 0 or x >= rows or y < 0 or y >= cols:
                continue
            elif (x, y) not in visited and mat[x][y] > mat[i][j] and mat[x][y] < 9:
                visited.add((x,y))
                dfs(x, y, visited)

    lows = findLowPoints(mat)
    for i, j in lows:
        visited = set([(i, j)])
        dfs(i, j, visited)
        sizes.append(len(visited))

    sizes.sort(reverse=True)
    print(sizes)

    return reduce(lambda x, y: x*y, sizes[:3])
```

```python
findTop3Basins(mat)
```
