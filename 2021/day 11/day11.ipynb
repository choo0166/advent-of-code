```python
import numpy as np
data = open(
    '/content/drive/MyDrive/Colab Notebooks/AOC/2021/day11.txt',
    'r')

mat = []
for line in data.readlines():
    arr = list(line.rstrip())
    mat.append(list(map(int, arr)))

mat = np.array(mat, dtype = np.int8)
```

## Part 1

```python
def findTotalFlashes(input: np.ndarray, steps: int) -> int:
    rows, cols = input.shape
    ans = 0

    def getNeighbors(i, j):
        neis = []
        # look at points in top, left, bottom, right, top
        # diagonal left/right, bottom diagonal left/right
        for x,y in [(i-1, j),
                    (i, j-1),
                    (i+1, j),
                    (i, j+1),
                    (i-1, j-1),
                    (i-1, j+1),
                    (i+1, j-1),
                    (i+1, j+1)
                    ]:
                    if x >= rows or x < 0 or y >= cols or y < 0:
                        continue
                    else:
                        neis.append((x,y))
        return neis

    for step in range(steps):
        input += 1
        flashed = set() # track which indexes already flashed
        f = np.argwhere(input > 9)
        while len(f) > 0:
            flashed.update([(ind[0], ind[1]) for ind in f])
            for ind in f:
                i, j = ind
                for nei in getNeighbors(i, j):
                    if nei not in flashed: input[nei] += 1
                input[i, j] = 0     # reset energy level
            f = np.argwhere(input > 9)

        ans += len(flashed)

    return ans
```

```python
findTotalFlashes(mat, 100)
```

## Part 2

```python
def findStepWithSimultFlashes(input: np.ndarray) -> int:
    rows, cols = input.shape
    step = 0
    def getNeighbors(i, j):
        neis = []
        # look at points in top, left, bottom, right, top
        # diagonal left/right, bottom diagonal left/right
        for x,y in [(i-1, j),
                    (i, j-1),
                    (i+1, j),
                    (i, j+1),
                    (i-1, j-1),
                    (i-1, j+1),
                    (i+1, j-1),
                    (i+1, j+1)
                    ]:
                    if x >= rows or x < 0 or y >= cols or y < 0:
                        continue
                    else:
                        neis.append((x,y))
        return neis

    while True:
        step += 1
        input += 1
        flashed = set() # track which indexes already flashed
        f = np.argwhere(input > 9)
        while len(f) > 0:
            flashed.update([(ind[0], ind[1]) for ind in f])
            for ind in f:
                i, j = ind
                for nei in getNeighbors(i, j):
                    if nei not in flashed: input[nei] += 1
                input[i, j] = 0     # reset energy level
            f = np.argwhere(input > 9)
        # if all energy levels are 0 then all have flashed
        if np.sum(input) == 0:
            return step
```
