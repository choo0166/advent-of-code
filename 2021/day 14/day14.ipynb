```python
# Build insertion rule mapping
rules = {}
for line in data[1].split("\n"):
    pair, insert = line.split(" -> ")
    rules[pair] = insert
```

## Part 1

```python
from typing import Dict
from collections import Counter

def findPairInsertionResultP1(template: str,
                              ruleMap: Dict[str, str],
                              steps: int) -> int:
    result = template
    print(f"Template: {result}")
    for step in range(steps):
        s = ""
        for i in range(len(result)-1):
            pair = result[i:i+2]
            if i == len(result) - 2:
                s += pair[0] + ruleMap.get(pair, "") + pair[1]
            else:
                s += pair[0] + ruleMap.get(pair, "")

        result = s
        print(f"After step {step + 1}: {result}")

    counts = Counter(result)
    return max(counts.values()) - min(counts.values())
```

```python
findPairInsertionResultP1(template, rules, 10)
```

## Part 2

```python
from collections import defaultdict

def findPairInsertionResultP2(template: str,
                              ruleMap: Dict[str, str],
                              steps: int) -> int:
    # represent polymer as dictionary of pair counts
    result = defaultdict(int)
    for i in range(len(template)-1):
        pair = template[i:i+2]
        result[pair] += 1

    for step in range(steps):
        # At each step, for every matching pair in the insertion rule, we add
        # new pairs according to the insertion rule i.e. for AB -> C rule, we
        # have {"AB": 3} -> {"AC": 3, "CB": 3}.
        s = defaultdict(int)
        for pair, count in result.items():
            if pair in ruleMap:
                insert1 = pair[0] + ruleMap[pair]
                insert2 = ruleMap[pair] + pair[1]
                s[insert1] += count
                s[insert2] += count
            else:
                s[pair] += count
            result = s

    # Add counts only for 1st character of each pair
    # and add 1 to the count for the last character
    # which will be unchanged from the template
    counts = defaultdict(int)
    for pair, count in result.items():
        counts[pair[0]] += count

    counts[template[-1]] += 1

    return max(counts.values()) - min(counts.values())
```

```python
findPairInsertionResultP2(template, rules, 40)
```
