## Part 1

```python
from typing import List

def findCountsP1(data: List[str]):
    ans = 0
    for line in data:
        signals, output = line.split(" | ")
        for p in output.split():
            # Digits 1, 4, 7, 8 have 2, 4, 3 and 7 segments resp.
            if len(p) in {2, 4, 3, 7}:
                ans += 1

    return ans
```

## Part 2

```python
from typing import List, Dict

def decode(signals: str) -> Dict[int, set]:
    """
    Helper method to decode signals to digits and
    returns the digit to signal mapping
    """
    digit_map = {} # map of digit -> signal
    five_segment_signals = [] # signals for digits 2,3,5
    six_segment_signals = [] # signals for digits 0,6,9

    for s in signals.split():
        length = len(s)
        if length == 2:
            digit_map[1] = set(s)
        elif length == 3:
            digit_map[7] = set(s)
        elif length == 4:
            digit_map[4] = set(s)
        elif length == 5:
            five_segment_signals.append(set(s))
        elif length == 6:
            six_segment_signals.append(set(s))
        elif length == 7:
            digit_map[8] = set(s)

    assert(len(five_segment_signals) == 3)
    assert(len(six_segment_signals) == 3)

    horz_segments = set.intersection(*five_segment_signals)
    # print(horz_segments)

    for s in six_segment_signals:
        # signal for digit 0 does not contain all horizontal segments
        if not(horz_segments.issubset(s)):
            digit_map[0] = s
        # signal for digit 9 contains all segments of digit 4
        elif digit_map[4] <= s:
            digit_map[9] = s
        else:
            digit_map[6] = s

    for s in five_segment_signals:
        # signal for digit 3 contains all segment of digit 1
        if digit_map[1] <= s:
            digit_map[3] = s
        # signal for digit 5 combined with signal for digit 1
        # is the signal for digit 9
        elif s.union(digit_map[1]) == digit_map[9]:
            digit_map[5] = s
        else:
            digit_map[2] = s

    return digit_map


def findSumP2(data: List[str]) -> int:
    sum = 0
    for line in data:
        signals, output = line.split(" | ")
        signal_map = {frozenset(v):k for k,v in decode(signals).items()}
        num = ''
        for p in output.split():
            num += str(signal_map[frozenset(p)])
        sum += int(num)

    return sum
```

```python
findSumP2(data)
```
