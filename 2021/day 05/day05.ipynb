## Part 1

```python
from typing import List
from collections import defaultdict

def getNumOverlapsP1(data: List[str]) -> int:
    counts = defaultdict(int)
    for line in data:
        seg = line.split(' -> ')
        coords = [tuple(map(int, v.split(','))) for v in seg]
        (x1, y1), (x2, y2) = coords[0], coords[1]
        if x1 == x2:
            sign = 1 if y2 - y1 > 0 else -1
            for y in range(y1, y2+sign, sign):
                counts[(x1, y)] += 1
        elif y1 == y2:
            sign = 1 if x2 - x1 > 0 else -1
            for x in range(x1, x2+sign, sign):
                counts[(x, y1)] += 1

    return sum(v > 1 for v in counts.values())
```

## Part 2

```python
def getNumOverlapsP2(data: List[str]) -> int:
    counts = defaultdict(int)
    for line in data:
        seg = line.split(' -> ')
        coords = [tuple(map(int, v.split(','))) for v in seg]
        (x1, y1), (x2, y2) = coords[0], coords[1]
        if x1 == x2:
            sign = 1 if y2 - y1 > 0 else -1
            for y in range(y1, y2+sign, sign):
                counts[(x1, y)] += 1
        elif y1 == y2:
            sign = 1 if x2 - x1 > 0 else -1
            for x in range(x1, x2+sign, sign):
                counts[(x, y1)] += 1
        else:
            x_dir, y_dir = x2 - x1, y2 - y1
            sign_x = int(x_dir/abs(x_dir))
            sign_y = int(y_dir/abs(y_dir))

            for (x,y) in zip(range(x1, x2+sign_x, sign_x),
                             range(y1, y2+sign_y, sign_y)):
                counts[(x,y)] += 1

    return sum(v > 1 for v in counts.values())
```

```python
getNumOverlapsP2(data)
```
