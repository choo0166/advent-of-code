# Day 1

## Part 1

```python
data = open(
    '/content/drive/MyDrive/Colab Notebooks/AOC/2021/day1.txt',
    'r').read().split("\n")
data = list(map(int, data))
```

```python
ans = 0
for i in range(1, len(data)):
    if data[i] > data[i-1]:
        ans += 1
print(ans)
```

## Part 2

```python
ans = 0
prev = sum(data[0:3])
curr = 0
for i in range(1, len(data)):
    curr = sum(data[i: i+3]) # slicing out of index returns empty list
    if curr > prev:
        ans += 1
    prev = curr
```
