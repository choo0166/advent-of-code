```python
from collections import defaultdict

# Build undirected graph as adjacency list
g = defaultdict(list)

for line in data.readlines():
    u, v = line.strip().split('-')
    g[u].append(v)
    g[v].append(u)

for k, v in g.items():
    print(k, v)
```

## Part 1

```python
from typing import List, Dict

def findNumPathsP1(g: Dict[str, List[str]]) -> int:
    path, result = [], []
    visited = set(["start"])
    def dfs(start, path, visited):
        path.append(start)
        if start == "end":
            result.append(path.copy())
        else:
            for nei in g[start]:
                if nei not in visited or nei.isupper():
                    visited.add(nei)
                    dfs(nei, path, visited)
        # Backtrack from last visited node
        path.pop()
        if start.islower(): visited.remove(start)

    dfs("start", path, visited)
    # print(result)

    return len(result)
```

```python
findNumPathsP1(g)
```

## Part 2

```python
from collections import defaultdict

def findNumPathsP2(g: Dict[str, List[str]]) -> int:
    path, result = [], []
    visited = defaultdict(int)  # track visits to small caves
    visited['start'] += 1
    def dfs(start, path, visited):
        path.append(start)
        if start == "end":
            result.append(path.copy())
        else:
            for nei in g[start]:
                if nei not in visited:
                    if nei.islower(): visited[nei] += 1
                    dfs(nei, path, visited)
                # if no other small cave except start is visited twice, allow revisit
                elif nei != "start" and 2 not in visited.values():
                    visited[nei] += 1
                    dfs(nei, path, visited)
        # Backtrack from last visited node
        path.pop()
        if start.islower():
            if visited[start] > 1: visited[start] -= 1
            else: del visited[start]

    dfs("start", path, visited)
    # print(result)

    return len(result)
```

```python
findNumPathsP2(g)
```
