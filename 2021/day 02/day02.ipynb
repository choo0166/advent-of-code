# Day 2

```python
data = open(
    '/content/drive/MyDrive/Colab Notebooks/AOC/2021/day2.txt',
    'r')
course = data.readlines()
```

## Part 1

```python
horz = 0
depth = 0
for dir in course:
    direction, step = dir.split(sep = " ")
    if direction == "forward":
        horz += int(step)
    elif direction == "down":
        depth += int(step)
    else:
        depth -= int(step)

print(horz*depth)
```

## Part 2

```python
horz = 0
depth = 0
aim = 0
for dir in course:
    direction, step = dir.split(sep = " ")
    if direction == "forward":
        horz += int(step)
        depth += aim*(int(step))
    elif direction == "down":
        aim += int(step)
    else:
        aim -= int(step)

print(horz*depth)
```
