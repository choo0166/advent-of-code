## Part 1

```python
from typing import List

def simulateP1(state: List[int], n: int) -> int:
    def simulate(timer):
        if timer == 0:
            state.append(9)
            return 6
        else:
            return timer-1

    for _ in range(n):
        state = list(map(simulate, state))

    return len(state)

simulateP1(input.copy(), 80)
```

## Part 2

```python
def simulateP2(state: List[int], n: int) -> int:
    counts = [0]*9  # counts of fish with timer i, i=0,1,2,...,8
    for timer in state:
        counts[timer] += 1

    print(f"Initial counts: {counts}")

    for _ in range(n):
        babies = counts[0]
        for i in range(len(counts)-1):
            counts[i] = counts[i+1]

        counts[6] += babies
        counts[8] = babies

    return sum(counts)


simulateP2(input.copy(), 256)
```
